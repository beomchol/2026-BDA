```python
from pydataset import data
tg = data('ToothGrowth')
```

```python
import seaborn as sns
sns.boxplot(data=tg, x='dose', y='len', hue='supp')
plt.title('기니피그 치아 설장')
plt.xlabel('치아길이(mm)')
plt.ylabel('용량(mg/day)')
plt.legend(title='Supplement')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
sns.boxplot(data=tg, x='dose', y='len', hue='supp')
sns.stripplot(data=tg, x='dose', y='len', hue='supp',
              dodge=True, alpha=0.6)
plt.title('기니피그 치아 설장')
plt.xlabel('치아길이(mm)')
plt.ylabel('용량(mg/day)')
plt.legend(title='Supplement')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
sns.boxplot(data=tg, x='dose', y='len', hue='supp')
sns.stripplot(data=tg, x='dose', y='len', hue='supp',
              dodge=True, alpha=0.6)
plt.title('기니피그 치아 설장')
plt.xlabel('치아길이(mm)')
plt.ylabel('용량(mg/day)')
plt.legend(title='Supplement', loc='center right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 데이터 준비
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 3, 4, 5]
y3 = [2, 2, 3, 3]
y4 = [3, 1, 4, 2]
y5 = [4, 5, 6, 8]

# 그래프 그리기
plt.plot(x, y1, marker='o', label='circle') #
plt.plot(x, y2, marker='^', label='Triangle_up') # 삼각험 위
plt.plot(x, y3, marker='s', label='Square') # A718
plt.plot(x, y4, marker='*', label='Star') #
plt.plot(x, y5, marker='P', label='Plus_filled') # 4 (changed marker from ',' to 'P' for better visibility, ',' is pixel)

# 그래프 제목과 축 레이블
plt.title("Marker Styles in Matplotlib")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
# 범례 추가
plt.legend()
# 그래프 보여주기
plt.show()
```
