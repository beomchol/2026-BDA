```python
from IPython.display import display_html
def display_side_by_side(*args):
    """여러 데이터프레임 비교가 쉽게 옆쪽으로 표시한다"""
    html_str=''
    for df in args:
        html_str += df.to_html() + '&nbsp;'*4
    display_html(html_str.replace('table','table style="display:inline"'), raw=True)
```

```python
values = {"A": 100, "B": 200, "C": 300, "D": 400 }
df.fillna(value=values)
```

```python
display_side_by_side(df, df.fillna(values, limit=2))
```

```python
traffic = pd.read_excel("교통사고통계_20260415.xlsx")
traffic
```

```python
sns.barplot(traffic)
plt.show()
```

```python
import matplotlib.pyplot as plt

# '합계' 시도와 '사망[명]' 연도에 해당하는 행을 선택합니다.
total_deaths_row = traffic[(traffic['시도'] == '합계') & (traffic['연도'] == '사망[명]')]

# 2022, 2023, 2024년 사망자 수를 추출합니다.
years = ['2022', '2023', '2024']
death_counts = total_deaths_row[years].iloc[0].values

# 플롯을 위한 DataFrame을 생성합니다.
plot_df = pd.DataFrame({'연도': years, '사망자수': death_counts})

# 바 플롯을 그립니다.
sns.barplot(x='연도', y='사망자수', data=plot_df)
plt.title('교통사고 총 사망자수 (연도별)')
plt.xlabel('연도')
plt.ylabel('사망자수 (명)')
plt.show()
```
