```python
import numpy as np
import pandas as pd

# 데이터 생성
distance = np.arange(1, 11)  # km
time = distance * 3 + np.random.randint(0, 5, size=10)  # 분 (대략 거리*3 + 랜덤)

# 요금 = 기본요금 + 거리요금 + 시간요금
fare = 2000 + distance * 1200 + time * 100 + np.random.randint(-200, 200, size=10)

# 데이터프레임 생성
df = pd.DataFrame({
    "distance": distance,
    "time": time,
    "fare": fare
})

# CSV 저장
df.to_csv("taxi_fare_with_time.csv", index=False)

print(df)
```
